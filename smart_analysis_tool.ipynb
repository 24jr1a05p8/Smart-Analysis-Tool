```python
import pandas as pd
df=pd.read_csv("IPL.csv")
df.head()
```

```python
df.describe()
```

```python
import plotly.express as px

top_batters = (
    df.groupby("batter")["runs_batter"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig = px.bar(
    top_batters,
    x="batter",
    y="runs_batter",
    title="Top 10 IPL Run Scorers"
)

fig.show()
```

```python
import plotly.express as px

top_bowlers = (
    df.groupby("bowler")["bowler_wicket"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig = px.bar(
    top_bowlers,
    x="bowler",
    y="bowler_wicket",
    title="Top 10 IPL Wicket Takers"
)

fig.show()
```

```python
season_matches = (
    df.groupby("season")["match_id"]
    .nunique()
    .reset_index()
)

fig = px.line(
    season_matches,
    x="season",
    y="match_id",
    title="Matches Played Per Season",
    markers=True
)

fig.show()
```

```python
team_wins = (
    df[['match_id', 'match_won_by']]
    .drop_duplicates()
    .groupby('match_won_by')
    .size()
    .reset_index(name='wins')
    .sort_values('wins', ascending=False)
    .head(10)
)

fig = px.pie(
    team_wins,
    names='match_won_by',
    values='wins',
    title='Top Winning Teams'
)

fig.show()
```
